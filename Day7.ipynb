```python
#wap to print sum using  function

num1=int(input("Enter the number: "))
num2=int(input("Enter the number: "))

add=lambda a,b:a+b
print("sum: ",add(num1,num2))
```

```python
#wap to find largest number 

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

largest = lambda x, y: x if x > y else y

print("Largest number is:", largest(a, b))
```

```python
#wap to print mul using  function
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

mul= lambda x,y: a*b
print("mul: ",mul(a,b))
```

```python
#celcius to farhenheit
degree = float(input("Enter first number: "))
conversion= lambda d: d*1.8 +32
print("Farhenheit: ",conversion(degree))
```

```python
#multiplication table 
num=int(input("Enter the number: "))
for i in range(1,11):
    mul=lambda a: a*i
    print("Table of ",num,":",mul(num))
```

```python
#cube
num=int(input("Enter the number: "))
cube= lambda a: a*a*a
print("Cube: ",cube(num))
```
