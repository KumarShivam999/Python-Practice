```python
#Reverse a string
name="Shivam"
#print(name[::-1])
rev_str=""
for char in name:
    rev_str=char+rev_str
print(rev_str)
```

```python
#Count number of vowels in a string
name=input("Enter the number")
count=0

for char in name:
    if char=='a' or char=='e' or char=='i' or char=='o' or char=='u':
        count=count+1
print(count)
```

```python
#check wheteher a string is palindrome 
name="nitin"
rev_str=""
for char in name:
    rev_str=char+rev_str
if rev_str==name:
    print("Palindrome")
else:
    print("not palindrome")
```
