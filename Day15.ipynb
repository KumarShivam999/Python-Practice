```python
#Create a Employee class with a private attribute __salary. Add a getter and a setter  the setter shoild reject negative salaries. Add a method apply_raise(percent) thst increases salary using the setter internally
class Employee:
    def __init__(self, salary):
        self.__salary = salary
    def get_salary(self):
        return self.__salary
    def set_salary(self, salary):
        if salary > 0:
            self.__salary = salary
        else:
            print("Invalid salary")

    # Increase salary
    def apply_raise(self, percent):
        new_salary = self.__salary + (self.__salary * percent / 100)
        self.set_salary(new_salary)


# Create object
emp = Employee(50000)

# Display salary
print("Salary:", emp.get_salary())

# Apply raise
emp.apply_raise(10)

print("After Raise:", emp.get_salary())

# Try negative salary
emp.set_salary(-5000)

```
