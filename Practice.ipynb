```python
#wap to Take a year as input and show it is leap year or not.
year=int(input("Enter the number: "))
if year%4==0 or year%4==0 and year%100!=0:
    print("leap year")
else:
    print("not leap year")
```

```python
#wap to Take marks (0–100) and print:

# 90-100 → A
# 80-89  → B
# 70-79  → C
# 60-69  → D
# Below 60 → Fail

num=int(input("Enter the number: "))
if num>=90 and num<=100:
    print("A")
elif num>=80 and num<=89:
    print("B")
elif num>=70 and num<=79:
    print("C")
elif num>=60 and num<=69:
    print("D")
else:
    print("fail")


```

```python
#wap for Simple Calculator
num1=int(input("Enter the number: "))
num2=int(input("Enter the number: "))
sy=input("Enter the symbol (eg:+,-,*,/,%): ")
if sy=='+':
    print(num1+num2)
elif sy=='*':
    print(num1*num2)
elif sy=='%':
    if num1!=0 and num1>num2 :
        print(num1%num2)
    else:
        print("Invalid")
elif sy=='/':
    if num1!=0 and num1>num2 :
        print(num1/num2)
    else:
        print("Invalid")
else:
    print(num1-num2)
```

```python
#wap to Character Check Take one character as input.Print whether it is:Alphabet Digit Special Character
char=input("Enter the character: ")
if char.isalpha():
    print("Alphabet")
elif char.isdigit():
    print("Digit")
else:
    print("special character")
```
