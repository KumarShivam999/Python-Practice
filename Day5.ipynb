```python
# --- 1. Define the basic math functions ---

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error (Cannot divide by zero)"
    return x / y

def modulus(x, y):
    return x % y

def exponent(x, y):
    return x ** y


# --- 2. Main program logic ---

print("--- Simple Calculator ---")
print("Operations: + , - , * , / , % , **")

# Take inputs from the user
num1 = float(input("Enter first number: "))
operation = input("Enter symbol (+, -, *, /, %, **): ")
num2 = float(input("Enter second number: "))

# --- 3. Check the symbol and call the correct function ---

if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = subtract(num1, num2)
elif operation == "*":
    result = multiply(num1, num2)
elif operation == "/":
    result = divide(num1, num2)
elif operation == "%":
    result = modulus(num1, num2)
elif operation == "**":
    result = exponent(num1, num2)
else:
    result = "Invalid Operator!"

# Print the final result
print("Result:", result)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    return a/b
def mod(a,b):
    return a%b
def sq(a,b):
    return a**b

a=int(input("Enter the number: "))
b=int(input("Enter the number: "))
operation=input("Enter the symbol: ")
if operation=='+':
    result=add(a,b)
elif operation=='-':
    result=sub(a,b)
elif operation=='*':
    result=multiply(a,b)
elif operation=='/':
    result=divide(a,b)
elif operation=='%':
    result=mod(a,b)
elif operation=='**':
    result=sq(a,b)
else:
    result='Invalid'
print("Result: ",result)
```
