```python
# Project:Student Marks Analysis System
# This walkthrough build a small realistic pipeline using only NUMPY
# Have a [articipants type each step along with you pause after each block to check output
# Step 1- Dataset Creation
# Steo 2- Average marks per student
# Step 3- Highest and Lowest marks
# Step 4- Standard deviation per Student
# Step 5- Ranking Students
# Step 6- Top Students(Above class average)
# Step 7-Subject wise analysis
# Step 8- percentage calculation
# Step 9- Visualization Ready data









import numpy as np
students = np.array(["Shivam","Riya","Priya","Utkarsh","Aditi"])
subjects = np.array(["Python","C","C++","Java","Java Script"])
marks = np.array([
    [85, 90, 78, 92, 88],
    [76, 81, 85, 79, 80],
    [90, 88, 92, 95, 91],
    [65, 70, 72, 68, 74],
    [88, 84, 80, 86, 90]
])
print("Students: ",students)
print("Subjects: ",subjects)
print("Marks: ",marks)
print(marks.shape)
print(marks.dtype)
print(marks.size)
print(marks.ndim)
average_marks=np.mean(marks,axis=0)
print("Average Marks:",average_marks)


for i in range(len(students)):
    print(students[i],":",average_marks[i],2)


highest_marks = np.max(marks, axis=1)
print(highest_marks)

lowest_marks = np.min(marks, axis=1)
print(lowest_marks)


for i in range(len(students)):
    print(students[i])
    print("Highest:", highest_marks[i])
    print("Lowest :", lowest_marks[i])


std_deviation = np.std(marks, axis=1)
print(std_deviation)



rank = np.argsort(average_marks)[::-1]

print("STUDENT LEADERBOARD\n")

for i in range(len(rank)):
    print("Rank", i + 1)
    print("Student :", students[rank[i]])
    print("Average :", round(average_marks[rank[i]], 2))
    print("-" * 35)
topper = rank[0]
print("\nTopper :", students[topper])
print("Average :", round(average_marks[topper], 2))

last = rank[-1]
print("\nLast Rank :", students[last])
print("Average :", round(average_marks[last], 2))


    
```
