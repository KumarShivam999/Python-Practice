```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\kumar\Downloads\student_performance_project.csv')
print(df)   #load the dataset 
```

```python
print(df.isnull().sum())  #Check number of null values in each row
```

```python
print(df.describe())  #describe the column 
```

```python
print(df.info)  #give information about column
```

```python
print(df.shape) #shape of dataset
```

```python
df["Math"] = df["Math"].fillna(df["Math"].mean())
df["Science"]=df["Science"].fillna(df["Science"].mean())
df["English"]=df["English"].fillna(df["English"].mean())
```

```python
print(df.isnull().sum()) #count number of null values
```

```python

print(df.duplicated())  # Check duplicate rows


df = df.drop_duplicates() #remove the duplcate rows

print("Number of duplicate rows:", df.duplicated().sum())  #count number of duplicate rows 
```

```python
filtered_student=df[(df["City"] == "Delhi") | (df["City"] == "Mumbai")] #filtering student who are from delhi or mumbai
print(filtered_student)
```

```python
print(df[["Math", "Science", "English"]].sum())  #sum of all marks of all subject
```

```python
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)
print(df["Total"].mean())
```

```python
df['Result']=df['Total'].map(lambda Total: 'Pass' if Total>=180 else 'Fail')
print(df["Result"])
```

```python
gender_city_analysis = df[(df["City"] == "Delhi") & (df["Gender"] == "M")]
print(gender_city_analysis)
```

```python
report = df.groupby(["Gender", "City"])["Total"].mean()
print(report)
```

```python
topper_list = df.sort_values(by="Total", ascending=False).head(5)
print(topper_list)
```
