```python
#sum of all digits
num=int(input("Enter the number"))
sum=0;
while(num!=0):
    rem=num%10
    sum=sum+rem
    num=num//10
print(sum)
    
    
```

```python
#palindrome
num=int(input("Enter the number"))
pal=num
sum=0
while num!=0:
    rem=num%10
    sum=sum*10+rem
    num=num//10
print(sum)
if(pal==sum):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#wap to print upto n number skip the number that is divisible by 4
num=int(input("Enter the number"))
i=1
while i<=num:
    if i%4!=0:
        print(i)
    i=i+1
```

```python
# nested loop 
# 0 0 
# 0 1
# 0 2
# 1 0
# 1 1
# 1 2
# 2 0
# 2 1 
# 2 2


num=int(input("Enter the number: "))
for i in range (0,num):
    for j in range (0,num):
        print(i,j)
    
```
