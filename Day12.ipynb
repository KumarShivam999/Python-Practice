```python
#Store the last 15 temperature recorded during a day and display the highest and lowest and average temperature
temp = []

for i in range(15):
    value = float(input("Enter temperature: "))
    temp.append(value)

print("Temperatures Recorded:")
print(temp)

highest = max(temp)

lowest = min(temp)

average = sum(temp) / len(temp)

print("\nHighest Temperature:", highest)
print("Lowest Temperature:", lowest)
print("Average Temperature:", average)

```

```python
#Store the unique programming language known by a class and then display them alphabetically

languages = set()
n = int(input("Enter number of programming languages: "))
for i in range(n):
    lang = input("Enter programming language: ")
    languages.add(lang)

for lang in sorted(languages):
    print(lang)
```
