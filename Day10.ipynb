```python
#Hospital Patient Monitoring
#concept:Traversal
#Heart rate list. Print patients with heart rate > 100BPM and count them.

heart_rates = []

n = int(input("Enter the number of patients: "))

for i in range(n):
    rate = int(input(f"Enter heart rate of patient {i+1}: "))
    heart_rates.append(rate)

count = 0
for rate in heart_rates:
    if rate > 100:
        print(rate, "BPM")
        count += 1

print("Total patients ", count)
```
