```python
# Create a function that can accept any number of details about a family member such as name,age,city,occupation,mobile number,etc., and display all the information provided.


def family_member(**details):
    print( "\nFamily Member Details:")
    for key, value in details.items():
        print (key, ":", value)


name = input("Enter Name: ")
age = input("Enter Age: ")
city = input("Enter City: ")
occupation = input("Enter Occupation: ")
mobile = input("Enter Mobile Number: ")


member=family_member(
    name=name,
    age=age,
    city=city,
    occupation=occupation,
    mobile=mobile
)
print(member)

```
